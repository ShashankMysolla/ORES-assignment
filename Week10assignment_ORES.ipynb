```python
import pandas as pd
```

```python
health_measures = pd.read_csv('/content/measures of health.csv')
```

```python
health_measures
```

```python
#dropping unneeded columns
health_measures.drop(['US_Unhealthy_Days','US_Health_Status'],axis=1,inplace=True)
```

```python
health_measures
```

```python
#changing Datatype
health_measures.dtypes
```

```python
health_measures['Unhealthy_Days'] = health_measures['Unhealthy_Days'].astype (int)
health_measures['Min_Unhealthy_Days'] = health_measures['Min_Unhealthy_Days'].astype (int)
health_measures['Max_Unhealthy_Days'] = health_measures['Max_Unhealthy_Days'].astype (int)
health_measures['CI_Min_Unhealthy_Days'] = health_measures['CI_Min_Unhealthy_Days'].astype (int)
health_measures['CI_Max_Unhealthy_Days'] = health_measures['CI_Max_Unhealthy_Days'].astype (int)
health_measures.dtypes
```

```python
health_measures.isna().sum()
```

```python
health_measures = health_measures.rename(columns={'Strata_ID_Number':'Strata_ID'})
```

```python
health_measures = health_measures.rename(columns={'CHSI_State_Name':'State'})
```

```python
health_measures.head()
```

```python
#calculated field of columns
health_measures['Total_Unhealthy_Days'] = health_measures['Min_Unhealthy_Days'] + health_measures['Max_Unhealthy_Days']
health_measures
```

```python
sum_by_state = health_measures.groupby('State').sum()
sum_by_state
```

```python
health_measures = health_measures.drop_duplicates()
health_measures
```

```python
q1 = health_measures['Unhealthy_Days'].quantile(0.25)
q3 = health_measures['Unhealthy_Days'].quantile(0.75)
iqr = q3-q1
```

```python
q1,q3,iqr
```

```python
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)
upper_limit,lower_limit
```

```python
new_data= health_measures.loc[(health_measures['Unhealthy_Days'] <  upper_limit) & (health_measures['Unhealthy_Days'] >  lower_limit)]
print('before',len(health_measures))
print('after',len(new_data))
print('outliers',len(health_measures)-len(new_data))
```
